```r
library(arules)

data(Mushroom)
```

```r
I <- apriori(Mushroom, parameter = list(target = "frequent itemsets", support = 0.1))
```

```r
R <- ruleInduction(I, Mushroom, confidence= 0.6, control = list(verbose = TRUE, method="apriori"))
```

```r
PR <- subset(R, subset = rhs %in% "Class=poisonous" & size(lhs)==1)
```

```r
inspect(PR)
```
